```python
%matplotlib inline
import scrublet as scr
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rc('font', size=14)
plt.rcParams['pdf.fonttype'] = 42
```

```python
input_dir = '/project/fsepru/PBMC5/Filtered/PBMC2/'
PBMC2_counts_matrix = scipy.io.mmread(input_dir + 'matrix.mtx').T.tocsc()
PBMC2_genes = np.array(scr.load_genes(input_dir + 'features.tsv', delimiter='\t', column=1))
print('Counts matrix shape: {} rows, {} columns'.format(PBMC2_counts_matrix.shape[0], PBMC2_counts_matrix.shape[1]))
print('Number of genes in gene list: {}'.format(len(PBMC2_genes)))
```

```python
PBMC2_scrub = scr.Scrublet(PBMC2_counts_matrix, expected_doublet_rate=0.06)
PBMC2_doublet_scores, PBMC2_predicted_doublets = PBMC2_scrub.scrub_doublets(min_counts=2, min_cells=3, min_gene_variability_pctl=85, n_prin_comps=30)
PBMC2_scrub.call_doublets(threshold=0.0)
PBMC2_scrub.plot_histogram();
```
