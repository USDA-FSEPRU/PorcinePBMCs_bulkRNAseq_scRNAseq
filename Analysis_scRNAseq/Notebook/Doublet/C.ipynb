```python
%matplotlib inline
import scrublet as scr
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rc('font', size=14)
plt.rcParams['pdf.fonttype'] = 42
```

```python
input_dir = '/project/fsepru/PBMC5/Filtered/C/'
C_counts_matrix = scipy.io.mmread(input_dir + 'matrix.mtx').T.tocsc()
C_genes = np.array(scr.load_genes(input_dir + 'features.tsv', delimiter='\t', column=1))
print('Counts matrix shape: {} rows, {} columns'.format(C_counts_matrix.shape[0], C_counts_matrix.shape[1]))
print('Number of genes in gene list: {}'.format(len(C_genes)))
```

```python
C_scrub = scr.Scrublet(C_counts_matrix, expected_doublet_rate=0.06)
C_doublet_scores, C_predicted_doublets = C_scrub.scrub_doublets(min_counts=2, min_cells=3, min_gene_variability_pctl=85, n_prin_comps=30)
C_scrub.call_doublets(threshold=0.0)
C_scrub.plot_histogram();
```
